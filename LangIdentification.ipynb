```python
# The source of the Data is Bible, the first three chapters of the book of Genesis
```

```python
# Read in the text from a file
```

```python
from nltk.tokenize import RegexpTokenizer
import string
import re
```

```python
# Reading Amharic Text from a file and Tokenizing it
```

```python
f = 'C:\Python3\LangIdentification\AmharicTranscript.txt'
lines = open(f, encoding="utf-8").readlines()
ETHIOPICNUMERALS='፩|፪|፫|፬|፭|፮|፯|፰|፱|፲|፳|፴|፵|፶|፷|፸|፹|፺|፻|፼'
ETHIOPICPUNCTUATIONS='፠|፡|።|፣|፤|፥|፦|፧|፨'
OTHER=r"a-z|A-Z|\d"
AmharicWords=[]
for line in lines:
    tokenizer=RegexpTokenizer(r'\w+')   
    line = re.sub(ETHIOPICNUMERALS,"", line)
    line = re.sub(ETHIOPICPUNCTUATIONS,"", line)
    line = line.strip(string.ascii_uppercase)
    line = line.strip(string.ascii_lowercase)
    line = line.strip(string.punctuation)
    line = line.strip(string.digits)
    line = line.strip(string.whitespace)
    line = line.strip('\ufeff')
    line = nltk.tokenize.wordpunct_tokenize(line)    
    for w in line:       
        AmharicWords.append(w)    
print(AmharicWords)
```

```python
# Reading Geez Text from a file and Tokenizing it
```

```python
f = 'C:\Python3\LangIdentification\GeezTranscript.txt'
lines = open(f, encoding="utf-8").readlines()

GeezWords=[]
for line in lines:
    tokenizer=RegexpTokenizer(r'\w+')   
    line = re.sub(ETHIOPICNUMERALS,"", line)
    line = re.sub(ETHIOPICPUNCTUATIONS,"", line)
    line = line.strip(string.ascii_uppercase)
    line = line.strip(string.ascii_lowercase)
    line = line.strip(string.punctuation)
    line = line.strip(string.digits)
    line = line.strip(string.whitespace)
    line = line.strip('\ufeff')
    line = nltk.tokenize.wordpunct_tokenize(line)    
    for w in line:       
        GeezWords.append(w)
        
print(GeezWords)
```

```python
# Reading Tigrigna Text from a file and Tokenizing it
```

```python
f = 'C:\Python3\LangIdentification\TigrignaTranscript.txt'
lines = open(f, encoding="utf-8").readlines()
TigrignaWords=[]

for line in lines:
    tokenizer=RegexpTokenizer(r'\w+')   
    line = re.sub(ETHIOPICNUMERALS,"", line)
    line = re.sub(ETHIOPICPUNCTUATIONS,"", line)
    line = line.strip(string.ascii_uppercase)
    line = line.strip(string.ascii_lowercase)
    line = line.strip(string.punctuation)
    line = line.strip(string.digits)
    line = line.strip(string.whitespace)
    line = line.strip('\ufeff')
    line = nltk.tokenize.wordpunct_tokenize(line)    
    for w in line:       
        TigrignaWords.append(w)
        
print(TigrignaWords)
```

```python
#Creating a Panda dataframe to store the text data with label
```

```python
import pandas as pd
from pandas import DataFrame
```

```python
dfT=pd.DataFrame(data=TigrignaWords, columns=["Tokens"])
dfT=dfT.assign(Language="Tigrigna")

dfA=pd.DataFrame(data=AmharicWords, columns=["Tokens"])
dfA=dfA.assign(Language="Amharic")

dfG=pd.DataFrame(data=GeezWords, columns=["Tokens"])
dfG=dfG.assign(Language="Geez")
```

```python
# Appending the dataframes into one big dataframe
```

```python
dfAll=pd.concat([dfA,dfG,dfT])
dfAll
```

```python
dfAll.Language.value_counts()
```

```python
dfAll['LanguageNum']=dfAll.Language.map({'Amharic':0,'Geez':1,'Tigrigna':2})
```

```python
X=dfAll.Tokens
y=dfAll.LanguageNum
print(X.shape)
print(y.shape)
```

```python
## Splitting Traing and Test data
```
